```python
# importing libraries- BeautifulSoup, requests
import pandas as pd
from bs4 import BeautifulSoup
import requests 
import warnings
warnings.filterwarnings('ignore')


url = "https://www.billboard.com/charts/hot-100"


response = requests.get(url)
```

```python
response.status_code
```

```python
# creating the soup
soup=BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
songs_all = soup.select('span.chart-element__information__song.text--truncate.color--primary')
songs_all
```

```python
songs_all = []
for p in songs_all: 
    songs_all.append((p.get_text()))
```

```python
songs_all
```

```python
len(songs_all)
```

```python
artist_all = soup.select('span.chart-element__information__artist.text--truncate.color--secondary')
artist_all
```

```python
len(artist_all)
```

```python
artists_all = []
for a in artist_all: 
    artists_all.append((a.get_text()))
```
